```python
import csv
from collections import defaultdict

def sum_marks(csv_file):
    # Initialize a dictionary to store the sum of marks for each roll number
    marks_dict = defaultdict(int)

    # Open the CSV file for reading
    with open(csv_file, mode='r') as file:
        reader = csv.reader(file)

        # Skip the header row
        next(reader)

        # Iterate over each row in the CSV file
        for row in reader:
            roll_no, marks = row
            # Convert marks to integer and add to the corresponding roll number in the dictionary
            marks_dict[roll_no] += int(marks)

    return marks_dict

def main():
    # Specify the path to the CSV file
    csv_file = '/content/Mini Project - Marks Adding.csv'  # Replace with your file path

    # Get the dictionary containing the sum of marks for each roll number
    result = sum_marks(csv_file)

    # Print the results
    for roll_no, total_marks in result.items():
        print(f"Roll No: {roll_no}, Total Marks: {total_marks}")

if __name__ == "__main__":
    main()
```
